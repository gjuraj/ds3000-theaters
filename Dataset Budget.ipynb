```python
import re
def create_url(base_url):
    """Increments the url given by replacing the digit at the end. 
    
    Args:
        base_url(String) : The url that will be used to increment to produce a new url.
                
    Returns:
        url (String): The new url. 
    """
    list_nums = re.findall('\d+', base_url)
    if list_nums:
        length_string = len(list_nums[0]) + 1
        num = int(list_nums[0])
        num += 100
    else:
        num = 101
        return f"{base_url}/{num}"
    return f"{base_url[:-length_string]}/{num}"
```

```python
create_url("https://www.the-numbers.com/movie/budgets/all")
```

```python
import urllib.request as urllib
from bs4 import BeautifulSoup

hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Connection': 'keep-alive'}

def get_soup(url):
    """Gets the soup of the url. 
    
    Args:
        url (String) : url used to get the soup object.
                
    Returns:
        soup (soup): The soup object for the image.
    """
    req = urllib.Request(url=url, headers=hdr) 
    html = urllib.urlopen(req)
    soup = BeautifulSoup(html.read())
    html.close()
    return soup
```

```python
get_soup("https://www.the-numbers.com/movie/budgets/all/1501")
```

```python
import pandas as pd
def create_dataframe_for_page(soup):
    """Creates a dataframe for the page. 
    
    Args:
        soup (soup object) : the soup object used to make the dataframe. 
                
    Returns:
        df (dataframe): The dataframe for the page. 
    """
    table = soup.find("table")
    list_movies = table.find_all("tr")
    list_names=[]
    list_budget=[]
    for movie in list_movies:
        href = movie.find_all("a")
        td = movie.find_all("td")
#         print(td)
        name = ""
        if href == []:
            continue
        else:
            name  = href[1].getText()
        budget = td[3].getText()[1:]
        list_names.append(name)
        list_budget.append(budget)
    df = pd.DataFrame(list(zip(list_names, list_budget)),
               columns =['Movie name', 'budget'])

    return df
```

```python
create_dataframe_for_page(get_soup("https://www.the-numbers.com/movie/budgets/all"))
```

```python
import pandas as pd
# used to go through all the web links and get a combined dataframe for the 6000+ movies. 
list_dict = []
starting_url = "https://www.the-numbers.com/movie/budgets/all"
i = 0
while (i< 61):
    soup = get_soup(starting_url)
    page_dict = create_dataframe_for_page(soup)
    list_dict.append(page_dict)
    starting_url = create_url(starting_url)
    print(starting_url)
    i+=1
    print(i)
dataset = pd.concat(list_dict)
print(dataset)
dataset.to_csv("budget_movies.csv" ,index=False)
```
